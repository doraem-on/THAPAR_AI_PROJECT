```python
# ‚îÄ‚îÄ Locate CSV files ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
import os, glob

# Try standard Kaggle input paths
SEARCH_DIRS = [
    '/kaggle/input/',
    '../input/',
    './',
]

csv_files = []
for d in SEARCH_DIRS:
    csv_files = glob.glob(os.path.join(d, '**', '*.csv'), recursive=True)
    if csv_files:
        print(f'üìÅ Found {len(csv_files)} CSV file(s) in {d}')
        break

if not csv_files:
    raise FileNotFoundError(
        'No CSV files found. Please add the CICIDS-2017 dataset via:\n'
        '  Notebook ‚Üí Add Data ‚Üí Search "CICIDS 2017" or "Network Intrusion Dataset"'
    )

for f in csv_files:
    print(f'   {f}')
```

## 3. üî¨ Data Quality Audit
> CICIDS-2017 is **known to contain Inf values and NaN** due to division-by-zero in CICFlowMeter (e.g., `Flow Bytes/s`). We handle all of these.

```python
df = df_raw.copy()

# ‚îÄ‚îÄ 1. Drop pure-string / ID columns ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
drop_candidates = ['Flow ID', 'Source IP', 'Destination IP', 'Timestamp',
                   'Source Port', 'Destination Port']   # these are identifiers, not features
drop_candidates = [c for c in drop_candidates if c in df.columns]
df.drop(columns=drop_candidates, inplace=True)
print(f'Dropped ID columns: {drop_candidates}')

# ‚îÄ‚îÄ 2. Replace Inf / -Inf with NaN ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
inf_count = np.isinf(df.select_dtypes(include=np.number)).sum().sum()
df.replace([np.inf, -np.inf], np.nan, inplace=True)
print(f'Inf values replaced : {inf_count:,}')

# ‚îÄ‚îÄ 3. Missing value audit ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
nan_cols = df.isnull().sum()
nan_cols = nan_cols[nan_cols > 0].sort_values(ascending=False)
print(f'\nColumns with NaN ({len(nan_cols)} total):')
print(nan_cols)

# Impute with column median (safe for skewed distributions)
num_cols = df.select_dtypes(include=np.number).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
print(f'\nNaN after imputation: {df.isnull().sum().sum()}')

# ‚îÄ‚îÄ 4. Duplicate rows ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
dupes = df.duplicated().sum()
print(f'Duplicate rows      : {dupes:,}  ‚Üí dropping...')
df.drop_duplicates(inplace=True)

print(f'\n‚úÖ Clean shape      : {df.shape}')
```

```python
# ‚îÄ‚îÄ Create Binary Label: Benign=0, Malicious=1 ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
df['attack_type']  = df[LABEL_COL].str.strip()     # keep original for deep-dive
df['binary_label'] = (df['attack_type'] != 'BENIGN').astype(int)

print('Binary class distribution:')
bc = df['binary_label'].value_counts()
print(f'  Benign    (0): {bc.get(0,0):>10,}  ({bc.get(0,0)/len(df)*100:.1f}%)')
print(f'  Malicious (1): {bc.get(1,0):>10,}  ({bc.get(1,0)/len(df)*100:.1f}%)')
```

```python
# ‚îÄ‚îÄ 4a. Class Distribution ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
fig = plt.figure(figsize=(18, 6))
gs  = gridspec.GridSpec(1, 3, figure=fig, wspace=0.35)

# Pie
ax0 = fig.add_subplot(gs[0])
sizes = [bc.get(0,0), bc.get(1,0)]
ax0.pie(sizes, labels=['Benign','Malicious'], colors=PALETTE,
        autopct='%1.1f%%', startangle=90,
        wedgeprops={'edgecolor':'#0d1117','linewidth':2},
        textprops={'color':'#e6edf3','fontsize':12})
ax0.set_title('Binary Class Split', pad=12)

# Attack type breakdown
ax1 = fig.add_subplot(gs[1:])
attack_counts = df[df['binary_label']==1]['attack_type'].value_counts()
colors_bar = plt.cm.plasma(np.linspace(0.15, 0.9, len(attack_counts)))
bars = ax1.barh(attack_counts.index, attack_counts.values,
                color=colors_bar, edgecolor='#21262d')
ax1.set_title('Attack Type Distribution (Malicious Traffic Only)')
ax1.set_xlabel('Flow Count')
for bar, v in zip(bars, attack_counts.values):
    ax1.text(v + max(attack_counts)*0.01, bar.get_y()+bar.get_height()/2,
             f'{v:,}', va='center', fontsize=8, color='#e6edf3')
ax1.invert_yaxis()

fig.suptitle('CICIDS-2017 ¬∑ Traffic Overview', fontsize=15, fontweight='bold', y=1.02)
plt.savefig('class_distribution.png', bbox_inches='tight', dpi=150)
plt.show()
```

<a id='5'></a>
## 5. üõ†Ô∏è Pre-processing Pipeline

```python
# ‚îÄ‚îÄ Drop low-variance & near-zero-variance features ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
from sklearn.feature_selection import VarianceThreshold

# Separate features from labels
feature_cols = [c for c in num_cols if c not in ['binary_label']]
X_raw = df[feature_cols].values
y     = df['binary_label'].values

# Low-variance filter (threshold = 0.01 after scaling)
vt = VarianceThreshold(threshold=0.01)
X_vt = vt.fit_transform(X_raw)
kept_cols = [feature_cols[i] for i, kept in enumerate(vt.get_support()) if kept]
print(f'Features before variance filter : {len(feature_cols)}')
print(f'Features after  variance filter : {len(kept_cols)}')
```

```python
# ‚îÄ‚îÄ Remove highly correlated features (> 0.97) ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
corr_df   = pd.DataFrame(X_vt, columns=kept_cols).corr().abs()
upper_tri = corr_df.where(np.triu(np.ones(corr_df.shape), k=1).astype(bool))
to_drop   = [c for c in upper_tri.columns if any(upper_tri[c] > 0.97)]
print(f'Highly correlated columns to drop ({len(to_drop)}): {to_drop[:10]}...')

final_cols = [c for c in kept_cols if c not in to_drop]
X_filtered = df[final_cols].values
print(f'\nFinal feature count after correlation pruning: {len(final_cols)}')
```

```python
# ‚îÄ‚îÄ Sample for speed (if dataset > 500k rows) ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
MAX_ROWS = 500_000
if len(df) > MAX_ROWS:
    # Stratified sample manually ‚Äî avoids pandas groupby.apply column-drop bug in newer versions
    benign_df    = df[df['binary_label'] == 0].sample(min((df['binary_label']==0).sum(), MAX_ROWS//2), random_state=42)
    malicious_df = df[df['binary_label'] == 1].sample(min((df['binary_label']==1).sum(), MAX_ROWS//2), random_state=42)
    df_sample    = pd.concat([benign_df, malicious_df], ignore_index=True)
    X_filtered   = df_sample[final_cols].values
    y            = df_sample['binary_label'].values
    print(f'Stratified sample: {len(df_sample):,} rows  (from {len(df):,})')
else:
    df_sample = df.copy().reset_index(drop=True)
    X_filtered = df_sample[final_cols].values
    y          = df_sample['binary_label'].values
    print(f'Using full dataset: {len(df):,} rows')

# ‚îÄ‚îÄ Train / Test Split ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
X_train, X_test, y_train, y_test = train_test_split(
    X_filtered, y, test_size=0.2, random_state=42, stratify=y
)

# ‚îÄ‚îÄ Scaling ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
scaler     = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc  = scaler.transform(X_test)

print(f'Train : {X_train_sc.shape}  |  Test: {X_test_sc.shape}')
```

<a id='7'></a>
## 7. ‚öñÔ∏è Handling Class Imbalance
> CICIDS-2017 has severe class imbalance. We use **SMOTE** (Synthetic Minority Over-sampling Technique) on the training set only ‚Äî never on test data.

```python
from collections import Counter

print(f'Before SMOTE: {Counter(y_train)}')

# Only apply SMOTE if significant imbalance exists
ratio = Counter(y_train)[1] / Counter(y_train)[0]
if ratio < 0.5:
    smote = SMOTE(random_state=42, n_jobs=-1)
    X_train_res, y_train_res = smote.fit_resample(X_train_sel, y_train)
    print(f'After SMOTE : {Counter(y_train_res)}')
else:
    X_train_res, y_train_res = X_train_sel, y_train
    print('Class balance acceptable ‚Äî SMOTE not applied.')
```

## 8. ü§ñ Model Training ‚Äî 10 Classifiers

```python
# ‚îÄ‚îÄ Define All Models ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
models = {
    'Logistic Regression'  : LogisticRegression(max_iter=1000, C=1.0, random_state=42, n_jobs=-1),
    'Decision Tree'        : DecisionTreeClassifier(max_depth=12, min_samples_split=10, random_state=42),
    'Random Forest'        : RandomForestClassifier(n_estimators=200, max_depth=15,
                                                    min_samples_split=10, n_jobs=-1, random_state=42),
    'Extra Trees'          : ExtraTreesClassifier(n_estimators=200, max_depth=15,
                                                  n_jobs=-1, random_state=42),
    'Gradient Boosting'    : GradientBoostingClassifier(n_estimators=150, max_depth=5,
                                                        learning_rate=0.1, random_state=42),
    'AdaBoost'             : AdaBoostClassifier(n_estimators=150, learning_rate=0.5, random_state=42),
    'KNN (k=7)'            : KNeighborsClassifier(n_neighbors=7, n_jobs=-1),
    'Naive Bayes'          : GaussianNB(),
}
if HAS_XGB:
    models['XGBoost'] = XGBClassifier(n_estimators=200, max_depth=6, learning_rate=0.1,
                                       use_label_encoder=False, eval_metric='logloss',
                                       n_jobs=-1, random_state=42)
if HAS_LGB:
    models['LightGBM'] = LGBMClassifier(n_estimators=200, max_depth=6, learning_rate=0.1,
                                         n_jobs=-1, random_state=42, verbose=-1)

print(f'Training {len(models)} models...')
```

```python
# ‚îÄ‚îÄ Train & Evaluate Each Model ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
results = {}
trained_models = {}

print(f'{"Model":<22} {"Acc":>7} {"Prec":>7} {"Rec":>7} {"F1":>7} {"AUC":>7} {"PR-AUC":>8} {"Time(s)":>8}')
print('‚îÄ' * 85)

for name, model in models.items():
    t0 = time.time()
    model.fit(X_train_res, y_train_res)
    elapsed = time.time() - t0

    y_pred  = model.predict(X_test_sel)
    y_proba = model.predict_proba(X_test_sel)[:,1] \
              if hasattr(model, 'predict_proba') else model.decision_function(X_test_sel)

    metrics = dict(
        Accuracy  = accuracy_score(y_test, y_pred),
        Precision = precision_score(y_test, y_pred, zero_division=0),
        Recall    = recall_score(y_test, y_pred, zero_division=0),
        F1        = f1_score(y_test, y_pred, zero_division=0),
        ROC_AUC   = roc_auc_score(y_test, y_proba),
        PR_AUC    = average_precision_score(y_test, y_proba),
        Time      = elapsed,
        y_pred    = y_pred,
        y_proba   = y_proba,
    )
    results[name]        = metrics
    trained_models[name] = model

    print(f'{name:<22} {metrics["Accuracy"]:>7.4f} {metrics["Precision"]:>7.4f} '
          f'{metrics["Recall"]:>7.4f} {metrics["F1"]:>7.4f} '
          f'{metrics["ROC_AUC"]:>7.4f} {metrics["PR_AUC"]:>8.4f} {elapsed:>8.1f}')

print('\n‚úÖ All models trained!')
```

```python
# ‚îÄ‚îÄ 9e. 10-Fold CV for Best Model ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
best_name  = board['F1'].idxmax()
best_model = trained_models[best_name]
print(f'üèÜ Best Model: {best_name}')
print(f'   Running 10-fold stratified CV...')

skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
cv_results = {}
for metric in ['accuracy','precision','recall','f1','roc_auc']:
    scores = cross_val_score(best_model, X_train_sel, y_train,
                              cv=skf, scoring=metric, n_jobs=-1)
    cv_results[metric] = scores
    print(f'   {metric:<12}: {scores.mean():.4f} ¬± {scores.std():.4f}')

# Plot
fig, axes = plt.subplots(1, 5, figsize=(20, 4))
fig.suptitle(f'10-Fold Cross-Validation ‚Äî {best_name}', fontsize=13, fontweight='bold')
colors_cv = [C_BENIGN, '#58a6ff', '#f0883e', C_MALICIOUS, '#d2a8ff']
for ax, (metric, scores), color in zip(axes, cv_results.items(), colors_cv):
    ax.boxplot(scores, patch_artist=True,
               boxprops={'facecolor':color,'alpha':0.7},
               medianprops={'color':'yellow','linewidth':2},
               whiskerprops={'color':'#8b949e'},
               capprops={'color':'#8b949e'},
               flierprops={'marker':'o','color':'#8b949e','alpha':0.5})
    ax.set_title(metric.title(), fontsize=10)
    ax.set_ylabel('Score')
    ax.set_ylim(max(0, scores.min()-0.02), 1.01)
    ax.text(1.15, scores.mean(), f'Œº={scores.mean():.4f}\nœÉ={scores.std():.4f}',
            color='#e6edf3', fontsize=8, va='center')
plt.tight_layout()
plt.savefig('cross_validation.png', bbox_inches='tight', dpi=150)
plt.show()
```

## 10. üß† SHAP Explainability
> **Why did the model flag this flow as malicious?** SHAP (SHapley Additive exPlanations) answers this ‚Äî per prediction, per feature.

```python
if HAS_SHAP:
    # Always use Random Forest for SHAP ‚Äî it's tree-based and guaranteed available
    # (Best model may be SVM / KNN / LR which don't support TreeExplainer)
    shap_model_name = 'Random Forest' if 'Random Forest' in trained_models else list(trained_models.keys())[0]
    # Prefer any tree-based model that was trained
    tree_models = [n for n in trained_models if hasattr(trained_models[n], 'feature_importances_')]
    shap_model_name = best_name if best_name in tree_models else (tree_models[0] if tree_models else None)

    if shap_model_name:
        shap_model = trained_models[shap_model_name]
        print(f'üß† Computing SHAP values using: {shap_model_name}')
        shap_sample_X = X_test_sel[:1000]

        explainer   = shap.TreeExplainer(shap_model)
        shap_values = explainer.shap_values(shap_sample_X)

        # shap_values is [class0, class1] for binary classifiers ‚Äî take class 1 (Malicious)
        sv = shap_values[1] if isinstance(shap_values, list) else shap_values

        # ‚îÄ‚îÄ SHAP Beeswarm Summary Plot ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
        plt.figure(figsize=(11, 7))
        shap.summary_plot(sv, shap_sample_X,
                          feature_names=sel_features,
                          plot_type='dot',
                          show=False,
                          max_display=15)
        plt.title(f'SHAP Beeswarm ‚Äî {shap_model_name}  (Malicious class)', fontsize=13, pad=12)
        plt.tight_layout()
        plt.savefig('shap_summary.png', bbox_inches='tight', dpi=150)
        plt.show()

        # ‚îÄ‚îÄ SHAP Bar (mean |SHAP|) ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
        plt.figure(figsize=(10, 6))
        shap.summary_plot(sv, shap_sample_X,
                          feature_names=sel_features,
                          plot_type='bar',
                          show=False,
                          max_display=15)
        plt.title(f'SHAP Feature Importance (Mean |SHAP|) ‚Äî {shap_model_name}', fontsize=12)
        plt.tight_layout()
        plt.savefig('shap_bar.png', bbox_inches='tight', dpi=150)
        plt.show()

        # ‚îÄ‚îÄ SHAP Waterfall for a single malicious prediction ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
        mal_idx = np.where(y_test[:1000] == 1)[0]
        if len(mal_idx) > 0:
            print(f'\nüîç Waterfall explanation for sample index {mal_idx[0]} (Malicious):')
            i = mal_idx[0]

            # shap_values shape varies by SHAP version:
            #   Old: list of arrays  [class0_arr, class1_arr]  ‚Üí sv already = class1 array (n_samples, n_features)
            #   New: single array    (n_samples, n_features, n_classes)
            raw_sv = explainer.shap_values(shap_sample_X[i:i+1])  # single row
            if isinstance(raw_sv, list):
                # list[class_idx] ‚Üí shape (1, n_features) ‚Üí squeeze to (n_features,)
                single_sv      = raw_sv[1][0]
                base_val       = explainer.expected_value[1]
            elif isinstance(raw_sv, np.ndarray) and raw_sv.ndim == 3:
                # shape (1, n_features, n_classes) ‚Üí class 1
                single_sv      = raw_sv[0, :, 1]
                base_val       = explainer.expected_value[1] if hasattr(explainer.expected_value, '__len__') \
                                 else explainer.expected_value
            else:
                # shape (1, n_features) ‚Äî already binary
                single_sv      = raw_sv[0]
                base_val       = explainer.expected_value if not hasattr(explainer.expected_value, '__len__') \
                                 else explainer.expected_value[0]

            exp = shap.Explanation(
                values        = single_sv,          # 1-D (n_features,)
                base_values   = float(base_val),
                data          = shap_sample_X[i],
                feature_names = sel_features
            )
            plt.figure(figsize=(10, 5))
            shap.plots.waterfall(exp, max_display=12, show=False)
            plt.title('SHAP Waterfall ‚Äî Single Malicious Flow', fontsize=12)
            plt.tight_layout()
            plt.savefig('shap_waterfall.png', bbox_inches='tight', dpi=150)
            plt.show()
    else:
        print('No tree-based model available for SHAP.')
else:
    # SHAP not installed ‚Äî fall back to standard feature importances
    print('‚ÑπÔ∏è  SHAP not installed ‚Äî showing sklearn feature importances instead.')
    tree_models = [n for n in trained_models if hasattr(trained_models[n], 'feature_importances_')]
    fi_model_name = best_name if best_name in tree_models else (tree_models[0] if tree_models else None)
    if fi_model_name:
        imp = pd.Series(trained_models[fi_model_name].feature_importances_,
                        index=sel_features).sort_values(ascending=True).tail(15)
        fig, ax = plt.subplots(figsize=(11, 6))
        bars = imp.plot(kind='barh', ax=ax,
                        color=plt.cm.plasma(np.linspace(0.2, 0.9, len(imp))))
        ax.set_title(f'Feature Importances ‚Äî {fi_model_name}', fontsize=13)
        ax.set_xlabel('Importance Score')
        for i, (val, name) in enumerate(zip(imp.values, imp.index)):
            ax.text(val + 0.001, i, f'{val:.4f}', va='center', fontsize=8, color='#e6edf3')
        plt.tight_layout()
        plt.savefig('feature_importances.png', bbox_inches='tight', dpi=150)
        plt.show()
```

## 11. üîç Attack-Type Deep Dive
> Beyond binary classification ‚Äî which specific **attack types** does our model miss?

```python
# Re-attach attack_type to the test rows for analysis
test_idx  = df_sample.index[int(len(df_sample)*0.8):]  # approximate, for analysis only

# Use the attack_type from the sample
# We'll rebuild predictions on the test portion with attack labels
df_test_portion = df_sample.tail(len(y_test)).copy()
df_test_portion['predicted_binary'] = results[best_name]['y_pred']
df_test_portion['true_binary']      = y_test
df_test_portion['correct']          = (df_test_portion['predicted_binary'] == 
                                        df_test_portion['true_binary'])

# Per-attack-type accuracy
attack_perf = df_test_portion.groupby('attack_type').agg(
    Total      = ('correct', 'count'),
    Correct    = ('correct', 'sum'),
).assign(Accuracy=lambda x: x.Correct / x.Total).sort_values('Accuracy')

print('Per-Attack-Type Detection Performance:')
print(attack_perf.to_string(float_format='%.4f'))

# Plot
fig, ax = plt.subplots(figsize=(13, 6))
colors_attack = [C_MALICIOUS if v < 0.95 else C_BENIGN
                 for v in attack_perf['Accuracy']]
bars = ax.barh(attack_perf.index, attack_perf['Accuracy'],
               color=colors_attack, edgecolor='#21262d')
ax.axvline(0.95, color='yellow', linestyle='--', lw=1.5, label='95% threshold')
ax.set_title(f'Per-Attack-Type Detection Accuracy ‚Äî {best_name}', fontsize=13)
ax.set_xlabel('Accuracy')
ax.set_xlim(0, 1.05)
ax.legend(fontsize=9)
for bar, val in zip(bars, attack_perf['Accuracy']):
    ax.text(min(val + 0.01, 1.01), bar.get_y() + bar.get_height()/2,
            f'{val:.3f}', va='center', fontsize=8, color='#e6edf3')
plt.tight_layout()
plt.savefig('attack_type_accuracy.png', bbox_inches='tight', dpi=150)
plt.show()
```

```python
# ‚îÄ‚îÄ False Negative Analysis: Which attacks sneak through? ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
fn_df = df_test_portion[
    (df_test_portion['true_binary'] == 1) &
    (df_test_portion['predicted_binary'] == 0)
]
fn_counts = fn_df['attack_type'].value_counts()

print(f'Total False Negatives (missed attacks): {len(fn_df):,}')
print('\nBreakdown by attack type:')
print(fn_counts.to_string())

if len(fn_counts) > 0:
    fig, ax = plt.subplots(figsize=(10, 4))
    fn_counts.plot(kind='barh', ax=ax, color=C_MALICIOUS, alpha=0.8)
    ax.set_title('‚ö†Ô∏è False Negatives (Missed Attacks) by Attack Type', fontsize=12)
    ax.set_xlabel('Count')
    ax.invert_yaxis()
    plt.tight_layout()
    plt.savefig('false_negatives.png', bbox_inches='tight', dpi=150)
    plt.show()
```

## üìù Key Takeaways

**1. Data Quality Matters More Than Model Choice**  
CICIDS-2017 contains thousands of Inf values and duplicate rows. Cleaning these had a larger impact than hyperparameter tuning.

**2. Recall > Precision in Cybersecurity**  
A missed attack (False Negative) is far more costly than a false alarm. We prioritise recall and use PR-AUC alongside ROC-AUC.

**3. SMOTE Helps But Has Limits**  
Synthetic oversampling improves detection of rare attack types, but very rare classes (Heartbleed, Infiltration) remain challenging.

**4. SHAP Reveals the "Why"**  
The most impactful features are flow-level statistics ‚Äî `Flow Duration`, `Flow Bytes/s`, `SYN Flag Count`, and `Packet Length` ‚Äî consistent with domain knowledge about network attacks.

**5. Ensemble Models Win**  
Tree-based ensembles (Random Forest, XGBoost, LightGBM) dominate. Logistic Regression and Naive Bayes underperform due to the non-linear nature of attack patterns.
